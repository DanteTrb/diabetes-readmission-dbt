```python
# ðŸ“¦ Import librerie
import duckdb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ðŸ“Œ Connessione al DB DuckDB
con = duckdb.connect("../data/diabetes.duckdb")

# ðŸ“¤ Query dal mart
df_hosp = con.execute("SELECT * FROM mart_hospital_usage").fetchdf()
df_hosp.head()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_hosp, x='readmitted', y='number_inpatient')
plt.title("Distribuzione dei ricoveri ospedalieri (inpatient) per tipo di readmission")
plt.xlabel("Readmission")
plt.ylabel("Numero di ricoveri inpatient")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_hosp, x='readmitted', y='number_outpatient')
plt.title("Visite ambulatoriali (outpatient) per tipo di readmission")
plt.xlabel("Readmission")
plt.ylabel("Numero di visite outpatient")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_hosp, x='readmitted', y='number_emergency')
plt.title("Accessi in pronto soccorso (emergency) per tipo di readmission")
plt.xlabel("Readmission")
plt.ylabel("Numero di accessi emergency")
plt.grid(True)
plt.tight_layout()
plt.show()
```
